# Automatización

En esta notebook pondremos a prueba el código que hemos desarrollado para efectuar el proceso de automatización, desde la generación de los insumos.

Supondremos que los insumos están listos el primer día hábil del mes.

```python
import datetime as dt


# La notebook se debera de ejecutar en el mismo mes en el cual se generan los insumos
# Fechas
dia_actual = dt.date.today()
inicio_mes_actual = str(dt.date(dia_actual.year, dia_actual.month, 1))
fin_mes_actual = str(dt.date(dia_actual.year, dia_actual.month + 1, 1) - dt.timedelta(days=1))

print("Inicio del mes actual: ", inicio_mes_actual)
print("Fin del mes actual: ", fin_mes_actual)
```

```python
# Realizamos la importacion de la libreria donde tenemos el codigo como
# clases
import Pym.m1.DataManip as pym

# Instanciamos para generar la informacion de julio
GIC = pym.GIC_BASE(f_init=inicio_mes_actual, f_fin=fin_mes_actual)
```

Notamos que lo anterior no requiere que nosotros modifiquemos algo, es decir, el código funciona de manera automática considerando solo el día en el que la ejecutaremos.

#### Simulación: obtención de información mensual

#### Consolidado mensual para el tablero

```python
# Creamos primero las bases de las rutas para mover el consolidado mensual
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio',
         'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

mes_act = dia_actual.month
mes_ant = dia_actual.month - 1
year = dia_actual.year
nombre_mes_ant = meses[mes_ant-1]

if mes_ant < 10:
    mes_ant = '0' + str(mes_ant)
else:
    mes_ant = str(mes_ant)
    
if mes_act < 10:
    mes_act = '0' + str(mes_act)
else:
    mes_act = str(mes_act)

nombre_ruta_act = mes_act + str(year)  
nombre_archivo = 'Consolidado' + nombre_mes_ant + str(year) + '.csv'
nombre_ruta = mes_ant + str(year)

print(nombre_archivo)
print(nombre_ruta)
print(nombre_ruta_act)
```

```python
# Creamos el consolidado del mes actual
GIC.consolidar_info_mens(m=nombre_ruta_act)
```

de modo que la información ya se encuentra lista para actualizar el tablero de Power BI. Mandaremos una notificación de escritorio para avisar que el proceso (mensual) ha terminado.

```python
from plyer import notification

notification.notify(
    title='Proceso terminado',
    message='Consolidado mensual creado, listo para pasar al tablero.')
```

### Consolidado trimestral

```python
# El consolidado trimestral lo debemos de ejecutar al mes siguiente de terminado el trimestre
trim = [0, 3, 6, 9] 
n_trim ={4: 1, 7: 2, 10: 3}
trimestre = dia_actual.month - 1 in  trim

# Realizamos el consolidado trimestral en dado caso de estar en el mes correcto
if trimestre:
    for i in n_trim.keys():
        if dia_actual.month == i:
            GIC.consolidar_info_trim(n_trim[dia_actual.month])

            # Enviamos la notificacion de escritorio
            notification.notify(
            title='Proceso terminado',
            message='Consolidado trimestral creado, listo para pasar al tablero.')
```
