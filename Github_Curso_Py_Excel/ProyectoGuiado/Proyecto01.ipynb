#  Diccionarios 

Queremos guardar la informaci贸n anterior de la siguiente manera:

* Clave-Precio: Para cada clave de producto queremos tener su precio asignado y guardado.
* Clave-Producto: Para cada clave de producto queremos tener su nombre asignado y guardado.

La estructura de _Diccionario_ en Python nos permite guardar informaci贸n as铆 como lo necesitamos pues nos permite guardar un valor asociado a una clave. La sintaxis es la siguiente:


```python
# lo siguente es un comentario multilinea
"""
diccionario = {clave1: valor1,
               clave2: valor2,
               clave3: valor3}
"""
```

De tal manera, en un diccionario pondremos como claves, justamente, las clave de los productos y como valores, el precio de dicho producto. De la misma manera crearemos un diccionario para guardar el nombre de los productos asociado a su clave. Antes de hacerlo veamos un ejemplo de un diccionario:

```python
# diccionario de numeros
nums = {'uno': 1,
        'dos': 2,
        'tres': 3}

# donde las claves del diccionario son las palabras y los valores son los numeros
# asi como en una lista, podemos acceder a los valores de un diccionario por medio de sus claves

print(nums['uno'])
print(nums['dos'])
print(nums['tres'])
```

```python
# otro ejemplo
lenguajes = {'Python': 'Lenguaje de programaci贸n de alto nivel',
             'Java': 'Lenguaje de programaci贸n de alto nivel',
              'C': 'Lenguaje de programaci贸n de medio nivel',
              'R': 'Lenguaje de programaci贸n de alto nivel'}

print(lenguajes['Python'])
print(lenguajes['C'])
```

Podemos agregar nuevos elementos a un diccionario de la siguiente manera:


```python
# vemos el diccionario completo
print(nums)
```

```python
# no tiene la clave, por ejemplo, 'cuatro', de modo que si escribirmos
nums['cuatro'] = 4

# estaremos agregando la clave 'cuatro' con el valor 4
nums
```

#  Generaci贸n de los datos 

```python
%%time
# Generaremos 1000 claves aleatorias para el diccionario
# Podemos crear diccionarios vacios que luego iremos llenando
clave_precio = {}
clave_producto = {}

# iteramos 1000 veces
for i in range(1000):
    #####################################################################################
    ######################### GENERAR CLAVE #############################################
    r1 = r.randint(0, longitud_abc)
    r2 = r.randint(0, longitud_abc)
    r3 = r.randint(0, longitud_abc)

    num1 = r.randint(0, 10)
    num2 = r.randint(0, 10)
    num3 = r.randint(0, 10)

    clave = abcdario[r1] + abcdario[r2] + abcdario[r3] + str(num1) + str(num2) + str(num3)
    #####################################################################################
    ######################### GENERAR PRECIO ############################################
    precio = r.random() * r.randint(1, 50000)
    precio = round(precio, 3)
    #####################################################################################
    ######################### ESCOGER UN PRODUCTO #######################################
    producto = r.choice(lineas)
    #####################################################################################
    ######################### LLENAR INFO EN LOS DICCIONARIO ############################
    clave_precio[clave] = precio
    clave_producto[clave] = producto
```

En vez de ver la informaci贸n como un diccionario, la veremos como un dataframe de pandas 

```python
# antes de ello veamos que hace la funcion items() de un diccionario
clave_precio.items()
```

cada tupla ser谩 una fila del dataframe

```python
# importacion necesaria para crear un dataframe
import pandas as pd

# con la funcion DataFrame() de la libreria pandas podemos crear un dataframe
df1 = pd.DataFrame(clave_precio.items(), columns=['clave', 'precio'])
df1.head()
```

```python
# Creamos otro diccionario
df2 = pd.DataFrame(clave_producto.items(), columns=['clave', 'producto'])
df2.head()
```

Nos gustar铆a tener un dataframe con la siguiente estructura:

en vez de tener dos dataframes, para lo cual podemos juntar los dataframes df1 y df2 tomando como criterio que la clave debe ser la misma

```python
# union de los dos dataframes tomando como criterio que la clave sea la misma
df_final = df1.merge(df2, on='clave')
df_final.head()
```

```python
# Reordenamos las columnas a como deseamos que queden
df_final = df_final[['clave', 'producto', 'precio']]

# vemos los ultimos 5 elementos del dataframe
df_final.tail()
```
